# Creating subset

```python
data_subset = data[['tree_id', 'tree_dbh','stump_diam',
       'curb_loc', 'status', 'health', 'spc_latin', 'steward', 'sidewalk', 'problems', 'root_stone',
       'root_grate', 'root_other', 'trunk_wire', 'trnk_light', 'trnk_other',
       'brch_light', 'brch_shoe', 'brch_other']]
data_subset
```

```python
#Viewing the trees with the dead status 
Deads = data_subset[data_subset['status']=='Dead']
Deads
```

```python
mask = ((data_subset['status']=='Stump') | (data_subset['status']=='Dead'))
```

```python
# Replacing NaN to Not Applicable to only health column
#data_subset.loc[mask, 'health'] = data_subset.loc[mask, 'health'].fillna('Not Applicable')
# Replacing NaN to Not Applicable to every columns
data_subset.loc[mask] = data_subset.loc[mask].fillna('Not Applicable')
```

```python
data_subset[data_subset['status']=='Stump']
```

```python
data_subset.isnull().sum()
```

```python
# Removing null values by replacing it following terms
data_subset['health'].fillna('None', inplace=True)
data_subset['spc_latin'].fillna('Good', inplace=True)
data_subset['sidewalk'].fillna('No Observation', inplace=True)
data_subset['problems'].fillna('No Damage', inplace=True)
```

```python
data_subset.isna().sum()
```

```python
big_trees = data_subset[(data_subset['tree_dbh']>60) | (data_subset['stump_diam']>60)]
big_trees
```

```python
data_subset = data_subset[(data_subset['tree_dbh']<=60) | (data_subset['stump_diam']<=60)]
data_subset
```

```python
# Separating stump and alive trees 
data_subset_alive = data_subset[data_subset['status']=='Alive']
data_subset_alive.groupby('spc_latin')['tree_dbh'].describe()
```
